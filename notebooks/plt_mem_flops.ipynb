```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 20
fontdict = {'family':'Times New Roman', 'style':'italic'}

plt.figure(figsize=(8*1.9,6))
res = [64, 128, 256, 512, 1024]

conv_mem = np.array([7.4, 197, 410, 1190, 4.340*1e3]) / 1e3
conv_flops = [0.439, 1.76, 7.03, 28.12, 112.5]

mamba_mem = np.array([14, 278, 764, 2.9*1e3, 11.539*1e3]) / 1e3
mamba_flops = [2.48, 9.836, 39.246, 156.9, 627.4]

attn_mem = np.array([28, 513, 12.9e3, 78e3]) / 1e3
attn_flops = np.array([5.89, 25.36, 178.3, 582.1])

swin_mem = np.array([12.9, 492.1, 1498, 9.391e3, 38.2e3]) / 1e3
swin_flops = np.array([3.29, 13.16, 78.92, 292.1, 1032.3])

plt.subplot(1,2,1)
plt.plot(res, conv_mem, marker='^', linestyle='--', lw=2, markersize=10, label='Conv')
plt.plot(res[:4], attn_mem, marker='v', linestyle='--', lw=2, markersize=10, label='Attention')
plt.plot(res, swin_mem, marker='o', linestyle='--', lw=2, markersize=10, label='Swin')
plt.plot(res, mamba_mem, marker='*', lw=2, markersize=12, label='LEVM')

plt.grid(linestyle='--')
plt.xticks([64, 128, 256, 512, 1024])
plt.xlabel('Resolution (px)')
plt.ylabel('GPU Memory (GB)')
plt.ylim([-1, 60])
plt.legend()

plt.subplot(1,2,2)
plt.plot(res, conv_flops, marker='^', linestyle='--', lw=2, markersize=10, label='Conv')
plt.plot(res[:4], attn_flops, marker='v', linestyle='--', lw=2, markersize=10, label='Attention')
plt.plot(res, swin_flops, marker='o', linestyle='--', lw=2, markersize=10, label='Swin')
plt.plot(res, mamba_flops, marker='*', lw=2, markersize=12, label='LEVM')

plt.grid(linestyle='--')
plt.xticks([64, 128, 256, 512, 1024])
plt.xlabel('Resolution (px)')
plt.ylabel('FLOPs (GB)')
plt.legend()
plt.gca().yaxis.set_label_coords(-0.08, 0.5)
plt.tight_layout()

# plt.subplots_adjust()

plt.savefig('mem_flops.png', dpi=300)
```

```python
methods = ['DiCNN', 'FusionNet', 'LAGConv', 'Invformer', 'DCFNet', 'HMPNet', 'PanDiff', 'PanMamba', 'Proposed']
latencies = [0.083, 0.065, 1.381, 1.691, 0.548, 0.741, 261.410, 0.781, 0.643]

methods = ['ResTFNet', 'SSRNet', 'HSRNet', 'MogDCN', 'Fusformer', 'DHIF', 'PSRT', '3DTNet', 'MIMO-SST', 'Proposed']
latencies = [0.485, 0.594, 0.690, 1.792, 12.143, 0.837, 0.621, 1.992, 1.321, 0.881]

# Set font to Times New Roman
plt.rcParams['font.family'] = 'Arial'

# Create bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Define colors, with a special color for "Proposed"
colors = ['skyblue'] * (len(methods) - 1) + ['lightcoral']

# Plot the bars and highlight the "Proposed" method
bars = ax.bar(methods, latencies, color=colors, edgecolor=['black']*(len(methods)-1) + ['darkred'])

# Highlight the "Proposed" bar with a white hatch pattern
bars[-1].set_hatch('//')
bars[-1].set_edgecolor('darkred')
bars[-1].set_linewidth(1.5)

# Set y-axis limit
ax.set_ylim(0, 2.2)

# Add labels and title
# ax.set_xlabel('Methods', fontsize=14)
ax.set_ylabel('Latency (seconds)', fontsize=16)
ax.set_xticklabels(methods, rotation=15, ha='right', fontsize=14)
# ax.set_title('Latency Comparison of Different Methods', fontsize=16)
# plt.xticks(rotation=15, ha='right')

# Add gridlines
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add time values above the bars, except for PanDiff
for bar, latency, method in zip(bars, latencies, methods):
    height = bar.get_height()
    if method == 'PanDiff' or method == 'Fusformer':
        ax.text(bar.get_x(), 2.1, f'{latency:.3f}', ha='left', va='center', fontsize=12, color='black')
    else:
        ax.text(bar.get_x() + bar.get_width() / 2, height, f'{latency:.3f}', ha='center', va='bottom', fontsize=12.7)

# Show the plot
plt.tight_layout()
plt.savefig('HISR-latency.pdf', dpi=200, bbox_inches='tight')
```
