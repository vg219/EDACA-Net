```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.patches import Rectangle

def analyze_and_visualize_test_dataset(h5_path):
    """
    ÂàÜÊûêÂπ∂ÂèØËßÜÂåñÊµãËØïÈõÜ‰∏≠ÊØèÂº†ÂõæÁöÑÊï∞ÊçÆËåÉÂõ¥
    """
    print(f"üîç ÂàÜÊûêÊï∞ÊçÆÈõÜ: {h5_path}")
    
    if not os.path.exists(h5_path):
        print(f"‚ùå Êñá‰ª∂‰∏çÂ≠òÂú®: {h5_path}")
        return
    
    # ÊâìÂºÄH5Êñá‰ª∂
    with h5py.File(h5_path, 'r') as f:
        print(f"üìä Êï∞ÊçÆÈõÜÂåÖÂê´ÁöÑÈîÆ: {list(f.keys())}")
        
        # Ëé∑ÂèñÊâÄÊúâÊï∞ÊçÆ
        data_dict = {}
        for key in f.keys():
            if isinstance(f[key], h5py.Dataset):
                data_dict[key] = f[key][:]
                print(f"  {key}: {f[key].shape} - {f[key].dtype}")
        
        # ÂàÜÊûêÊØè‰∏™Êï∞ÊçÆÁ±ªÂûã
        for data_name, data in data_dict.items():
            print(f"\n{'='*60}")
            print(f"üìà ÂàÜÊûê {data_name}")
            print(f"{'='*60}")
            
            if len(data.shape) == 4:  # (N, C, H, W)
                n_samples, n_channels, height, width = data.shape
                print(f"Êï∞ÊçÆÂΩ¢Áä∂: {n_samples}Âº†Âõæ √ó {n_channels}ÈÄöÈÅì √ó {height}√ó{width}")
                
                # ÂàÜÊûêÊØèÂº†ÂõæÁöÑÊï∞ÊçÆËåÉÂõ¥
                print(f"\nüìã ÊØèÂº†ÂõæÁöÑÊï∞ÊçÆËåÉÂõ¥ÂàÜÊûê:")
                print(f"{'ÂõæÁâáÂ∫èÂè∑':<8} {'ÊúÄÂ∞èÂÄº':<12} {'ÊúÄÂ§ßÂÄº':<12} {'ÂùáÂÄº':<12} {'Ê†áÂáÜÂ∑Æ':<12} {'Êï∞ÊçÆËåÉÂõ¥Á±ªÂûã'}")
                print("-" * 80)
                
                range_types = []
                
                for i in range(n_samples):
                    img = data[i]  # (C, H, W)
                    min_val = float(img.min())
                    max_val = float(img.max())
                    mean_val = float(img.mean())
                    std_val = float(img.std())
                    
                    # Âà§Êñ≠Êï∞ÊçÆËåÉÂõ¥Á±ªÂûã
                    if min_val < 0:
                        if max_val <= 1:
                            range_type = "ÊúâË¥üÂÄº[<0,‚â§1]"
                        elif max_val <= 255:
                            range_type = "ÊúâË¥üÂÄº[<0,‚â§255]"
                        else:
                            range_type = f"ÊúâË¥üÂÄº[<0,{max_val:.2f}]"
                    elif min_val >= 0 and max_val <= 1.1:
                        range_type = "[0,1]ËåÉÂõ¥"
                    elif min_val >= 0 and max_val <= 255:
                        if max_val > 100:  # Â§ßÊ¶ÇÁéáÊòØ[0,255]
                            range_type = "[0,255]ËåÉÂõ¥"
                        else:
                            range_type = f"[0,{max_val:.2f}]"
                    else:
                        range_type = f"ÂÖ∂‰ªñ[{min_val:.2f},{max_val:.2f}]"
                    
                    range_types.append(range_type)
                    
                    print(f"{i:<8} {min_val:<12.6f} {max_val:<12.6f} {mean_val:<12.6f} {std_val:<12.6f} {range_type}")
                
                # ÁªüËÆ°Êï∞ÊçÆËåÉÂõ¥Á±ªÂûã
                from collections import Counter
                range_counter = Counter(range_types)
                print(f"\nüìä Êï∞ÊçÆËåÉÂõ¥Á±ªÂûãÁªüËÆ°:")
                for range_type, count in range_counter.items():
                    print(f"  {range_type}: {count}Âº†Âõæ ({count/n_samples*100:.1f}%)")
                
                # ÂèØËßÜÂåñÊØèÂº†Âõæ
                visualize_images(data, data_name, range_types)
            
            else:
                print(f"‚ö†Ô∏è Ë∑≥ËøáÈùû4DÊï∞ÊçÆ: {data.shape}")

def visualize_images(data, data_name, range_types):
    """
    ÂèØËßÜÂåñÊØèÂº†ÂõæÁâá
    """
    n_samples, n_channels, height, width = data.shape
    print(f"\nüé® ÂèØËßÜÂåñ {data_name} ({n_samples}Âº†Âõæ)")
    
    # ËÆ°ÁÆóÁΩëÊ†ºÂ∏ÉÂ±Ä
    cols = min(4, n_samples)
    rows = (n_samples + cols - 1) // cols
    
    fig, axes = plt.subplots(rows, cols, figsize=(cols*4, rows*4))
    if rows == 1 and cols == 1:
        axes = [axes]
    elif rows == 1 or cols == 1:
        axes = axes.flatten()
    else:
        axes = axes.flatten()
    
    for i in range(n_samples):
        ax = axes[i] if i < len(axes) else None
        if ax is None:
            break
            
        img = data[i]  # (C, H, W)
        
        # Ê†πÊçÆÈÄöÈÅìÊï∞ÈÄâÊã©ÂèØËßÜÂåñÊñπÂºè
        if n_channels == 1:
            # ÁÅ∞Â∫¶Âõæ
            display_img = img[0]  # (H, W)
            im = ax.imshow(display_img, cmap='gray')
        elif n_channels == 3:
            # RGBÂõæÂÉèÔºåËΩ¨Êç¢‰∏∫(H, W, C)
            display_img = img.transpose(1, 2, 0)  # (H, W, 3)
            
            # Êï∞ÊçÆËåÉÂõ¥ÂΩí‰∏ÄÂåñÁî®‰∫éÊòæÁ§∫
            if display_img.max() > 1.1:
                # ÂèØËÉΩÊòØ[0,255]ËåÉÂõ¥
                display_img_norm = np.clip(display_img / 255.0, 0, 1)
            else:
                # ÂèØËÉΩÊòØ[0,1]ËåÉÂõ¥ÊàñÊúâË¥üÂÄº
                if display_img.min() < 0:
                    # ÊúâË¥üÂÄºÔºåÂÖàÂπ≥ÁßªÂÜçÂΩí‰∏ÄÂåñ
                    display_img_norm = display_img - display_img.min()
                    if display_img_norm.max() > 0:
                        display_img_norm = display_img_norm / display_img_norm.max()
                else:
                    display_img_norm = np.clip(display_img, 0, 1)
            
            im = ax.imshow(display_img_norm)
        else:
            # Â§öÂÖâË∞±ÔºåÈÄâÊã©Ââç3‰∏™ÈÄöÈÅì‰Ωú‰∏∫RGB
            if n_channels >= 3:
                display_img = img[:3].transpose(1, 2, 0)  # ÂèñÂâç3ÈÄöÈÅì
            else:
                display_img = img[0]  # Âè™ÊòæÁ§∫Á¨¨‰∏ÄÈÄöÈÅì
            
            # ÂΩí‰∏ÄÂåñÊòæÁ§∫
            if display_img.max() > 1.1:
                display_img_norm = np.clip(display_img / 255.0, 0, 1)
            else:
                if display_img.min() < 0:
                    display_img_norm = display_img - display_img.min()
                    if display_img_norm.max() > 0:
                        display_img_norm = display_img_norm / display_img_norm.max()
                else:
                    display_img_norm = np.clip(display_img, 0, 1)
            
            if len(display_img_norm.shape) == 3:
                im = ax.imshow(display_img_norm)
            else:
                im = ax.imshow(display_img_norm, cmap='viridis')
        
        # Ê∑ªÂä†Ê†áÈ¢òÂíå‰ø°ÊÅØ
        min_val = img.min()
        max_val = img.max()
        ax.set_title(f'Âõæ{i}: {range_types[i]}\n[{min_val:.3f}, {max_val:.3f}]', 
                    fontsize=10)
        ax.axis('off')
        
        # Ê†πÊçÆÊï∞ÊçÆËåÉÂõ¥Á±ªÂûãÊ∑ªÂä†ÂΩ©Ëâ≤ËæπÊ°Ü
        if "Ë¥üÂÄº" in range_types[i]:
            color = 'red'
        elif "[0,1]" in range_types[i]:
            color = 'green'
        elif "[0,255]" in range_types[i]:
            color = 'blue'
        else:
            color = 'orange'
        
        # Ê∑ªÂä†ÂΩ©Ëâ≤ËæπÊ°Ü
        rect = Rectangle((0, 0), width-1, height-1, linewidth=3, 
                        edgecolor=color, facecolor='none')
        ax.add_patch(rect)
        
        # Ê∑ªÂä†È¢úËâ≤Êù°
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    
    # ÈöêËóèÂ§ö‰ΩôÁöÑÂ≠êÂõæ
    for i in range(n_samples, len(axes)):
        axes[i].axis('off')
    
    plt.suptitle(f'{data_name} - ÊØèÂº†ÂõæÂèØËßÜÂåñ\nÁ∫¢Ê°Ü=ÊúâË¥üÂÄº, ÁªøÊ°Ü=[0,1], ËìùÊ°Ü=[0,255], Ê©ôÊ°Ü=ÂÖ∂‰ªñ', 
                fontsize=14)
    plt.tight_layout()
    plt.show()

def check_data_consistency(h5_path):
    """
    Ê£ÄÊü•Êï∞ÊçÆ‰∏ÄËá¥ÊÄßÈóÆÈ¢ò
    """
    print(f"\nüîç Êï∞ÊçÆ‰∏ÄËá¥ÊÄßÊ£ÄÊü•")
    print("="*60)
    
    with h5py.File(h5_path, 'r') as f:
        # Ê£ÄÊü•ÊâÄÊúâÊï∞ÊçÆÈõÜÁöÑÊï∞ÊçÆËåÉÂõ¥
        issues = []
        
        for key in f.keys():
            if isinstance(f[key], h5py.Dataset):
                data = f[key][:]
                if len(data.shape) == 4:
                    min_vals = []
                    max_vals = []
                    
                    for i in range(data.shape[0]):
                        min_vals.append(data[i].min())
                        max_vals.append(data[i].max())
                    
                    min_vals = np.array(min_vals)
                    max_vals = np.array(max_vals)
                    
                    # Ê£ÄÊü•ÊòØÂê¶ÊúâË¥üÂÄº
                    negative_count = np.sum(min_vals < 0)
                    if negative_count > 0:
                        issues.append(f"{key}: {negative_count}Âº†ÂõæÊúâË¥üÂÄº")
                    
                    # Ê£ÄÊü•Êï∞ÊçÆËåÉÂõ¥‰∏ç‰∏ÄËá¥
                    range_spread = max_vals.max() - min_vals.min()
                    if range_spread > 255:
                        issues.append(f"{key}: Êï∞ÊçÆËåÉÂõ¥ËøáÂ§ß ({min_vals.min():.3f} Âà∞ {max_vals.max():.3f})")
                    
                    # Ê£ÄÊü•Ê∑∑ÂêàÊï∞ÊçÆËåÉÂõ¥
                    has_0_1 = np.any((min_vals >= 0) & (max_vals <= 1.1))
                    has_0_255 = np.any(max_vals > 100)
                    if has_0_1 and has_0_255:
                        issues.append(f"{key}: Ê∑∑ÂêàÊï∞ÊçÆËåÉÂõ¥ (Êó¢Êúâ[0,1]ÂèàÊúâ[0,255]ËåÉÂõ¥)")
        
        if issues:
            print("‚ö†Ô∏è ÂèëÁé∞ÁöÑÈóÆÈ¢ò:")
            for issue in issues:
                print(f"  ‚Ä¢ {issue}")
        else:
            print("‚úÖ Êú™ÂèëÁé∞ÊòéÊòæÁöÑÊï∞ÊçÆ‰∏ÄËá¥ÊÄßÈóÆÈ¢ò")

# ‰∏ªÂàÜÊûêÂáΩÊï∞
def main():
    h5_path = '/data2/users/yujieliang/exps/Efficient-MIF-back-master-6-feat/data/ASSR/Botswana/datasets/Botswana_test_crop256x1280_patch256.h5'
    
    print("üîç ÂºÄÂßãÂàÜÊûêÊµãËØïÊï∞ÊçÆÈõÜ")
    print("="*80)
    
    # ÂàÜÊûêÂπ∂ÂèØËßÜÂåñ
    analyze_and_visualize_test_dataset(h5_path)
    
    # Êï∞ÊçÆ‰∏ÄËá¥ÊÄßÊ£ÄÊü•
    check_data_consistency(h5_path)
    
    print("\n" + "="*80)
    print("üèÅ ÂàÜÊûêÂÆåÊàê!")

# ËøêË°åÂàÜÊûê
main()
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.patches import Rectangle

def analyze_and_visualize_test_dataset(h5_path):
    """
    ÂàÜÊûêÂπ∂ÂèØËßÜÂåñÊµãËØïÈõÜ‰∏≠ÊØèÂº†ÂõæÁöÑÊï∞ÊçÆËåÉÂõ¥
    """
    print(f"üîç ÂàÜÊûêÊï∞ÊçÆÈõÜ: {h5_path}")
    
    if not os.path.exists(h5_path):
        print(f"‚ùå Êñá‰ª∂‰∏çÂ≠òÂú®: {h5_path}")
        return
    
    # ÊâìÂºÄH5Êñá‰ª∂
    with h5py.File(h5_path, 'r') as f:
        print(f"üìä Êï∞ÊçÆÈõÜÂåÖÂê´ÁöÑÈîÆ: {list(f.keys())}")
        
        # Ëé∑ÂèñÊâÄÊúâÊï∞ÊçÆ
        # Ëé∑ÂèñÊâÄÊúâÊï∞ÊçÆ (Âè™ÂèñÂâç10Âº†)
        data_dict = {}
        for key in f.keys():
            if isinstance(f[key], h5py.Dataset):
                # Âè™Âä†ËΩΩÂâç10Âº†Âõæ
                if len(f[key].shape) == 4:  # (N, C, H, W)
                    data_dict[key] = f[key][:10]  # Âè™ÂèñÂâç10Âº†
                    print(f"  {key}: {f[key].shape} -> Âä†ËΩΩÂâç10Âº†: {data_dict[key].shape} - {f[key].dtype}")
                else:
                    data_dict[key] = f[key][:]  # Èùû4DÊï∞ÊçÆÂÖ®ÈÉ®Âä†ËΩΩ
                    print(f"  {key}: {f[key].shape} - {f[key].dtype}")
        
        # ÂàÜÊûêÊØè‰∏™Êï∞ÊçÆÁ±ªÂûã
        for data_name, data in data_dict.items():
            print(f"\n{'='*60}")
            print(f"üìà ÂàÜÊûê {data_name}")
            print(f"{'='*60}")
            
            if len(data.shape) == 4:  # (N, C, H, W)
                n_samples, n_channels, height, width = data.shape
                print(f"Êï∞ÊçÆÂΩ¢Áä∂: {n_samples}Âº†Âõæ √ó {n_channels}ÈÄöÈÅì √ó {height}√ó{width}")
                
                # ÂàÜÊûêÊØèÂº†ÂõæÁöÑÊï∞ÊçÆËåÉÂõ¥
                print(f"\nüìã ÊØèÂº†ÂõæÁöÑÊï∞ÊçÆËåÉÂõ¥ÂàÜÊûê:")
                print(f"{'ÂõæÁâáÂ∫èÂè∑':<8} {'ÊúÄÂ∞èÂÄº':<12} {'ÊúÄÂ§ßÂÄº':<12} {'ÂùáÂÄº':<12} {'Ê†áÂáÜÂ∑Æ':<12} {'Êï∞ÊçÆËåÉÂõ¥Á±ªÂûã'}")
                print("-" * 80)
                
                range_types = []
                
                for i in range(n_samples):
                    img = data[i]  # (C, H, W)
                    min_val = float(img.min())
                    max_val = float(img.max())
                    mean_val = float(img.mean())
                    std_val = float(img.std())
                    
                    # Âà§Êñ≠Êï∞ÊçÆËåÉÂõ¥Á±ªÂûã
                    if min_val < 0:
                        if max_val <= 1:
                            range_type = "ÊúâË¥üÂÄº[<0,‚â§1]"
                        elif max_val <= 255:
                            range_type = "ÊúâË¥üÂÄº[<0,‚â§255]"
                        else:
                            range_type = f"ÊúâË¥üÂÄº[<0,{max_val:.2f}]"
                    elif min_val >= 0 and max_val <= 1.1:
                        range_type = "[0,1]ËåÉÂõ¥"
                    elif min_val >= 0 and max_val <= 255:
                        if max_val > 100:  # Â§ßÊ¶ÇÁéáÊòØ[0,255]
                            range_type = "[0,255]ËåÉÂõ¥"
                        else:
                            range_type = f"[0,{max_val:.2f}]"
                    else:
                        range_type = f"ÂÖ∂‰ªñ[{min_val:.2f},{max_val:.2f}]"
                    
                    range_types.append(range_type)
                    
                    print(f"{i:<8} {min_val:<12.6f} {max_val:<12.6f} {mean_val:<12.6f} {std_val:<12.6f} {range_type}")
                
                # ÁªüËÆ°Êï∞ÊçÆËåÉÂõ¥Á±ªÂûã
                from collections import Counter
                range_counter = Counter(range_types)
                print(f"\nüìä Êï∞ÊçÆËåÉÂõ¥Á±ªÂûãÁªüËÆ°:")
                for range_type, count in range_counter.items():
                    print(f"  {range_type}: {count}Âº†Âõæ ({count/n_samples*100:.1f}%)")
                
                # ÂèØËßÜÂåñÊØèÂº†Âõæ
                visualize_images(data, data_name, range_types)
            
            else:
                print(f"‚ö†Ô∏è Ë∑≥ËøáÈùû4DÊï∞ÊçÆ: {data.shape}")

def visualize_images(data, data_name, range_types):
    """
    ÂèØËßÜÂåñÊØèÂº†ÂõæÁâá
    """
    n_samples, n_channels, height, width = data.shape
    print(f"\nüé® ÂèØËßÜÂåñ {data_name} ({n_samples}Âº†Âõæ)")
    
    # ËÆ°ÁÆóÁΩëÊ†ºÂ∏ÉÂ±Ä
    cols = min(4, n_samples)
    rows = (n_samples + cols - 1) // cols
    
    fig, axes = plt.subplots(rows, cols, figsize=(cols*4, rows*4))
    if rows == 1 and cols == 1:
        axes = [axes]
    elif rows == 1 or cols == 1:
        axes = axes.flatten()
    else:
        axes = axes.flatten()
    
    for i in range(n_samples):
        ax = axes[i] if i < len(axes) else None
        if ax is None:
            break
            
        img = data[i]  # (C, H, W)
        
        # Ê†πÊçÆÈÄöÈÅìÊï∞ÈÄâÊã©ÂèØËßÜÂåñÊñπÂºè
        if n_channels == 1:
            # ÁÅ∞Â∫¶Âõæ
            display_img = img[0]  # (H, W)
            im = ax.imshow(display_img, cmap='gray')
        elif n_channels == 3:
            # RGBÂõæÂÉèÔºåËΩ¨Êç¢‰∏∫(H, W, C)
            display_img = img.transpose(1, 2, 0)  # (H, W, 3)
            
            # Êï∞ÊçÆËåÉÂõ¥ÂΩí‰∏ÄÂåñÁî®‰∫éÊòæÁ§∫
            if display_img.max() > 1.1:
                # ÂèØËÉΩÊòØ[0,255]ËåÉÂõ¥
                display_img_norm = np.clip(display_img / 255.0, 0, 1)
            else:
                # ÂèØËÉΩÊòØ[0,1]ËåÉÂõ¥ÊàñÊúâË¥üÂÄº
                if display_img.min() < 0:
                    # ÊúâË¥üÂÄºÔºåÂÖàÂπ≥ÁßªÂÜçÂΩí‰∏ÄÂåñ
                    display_img_norm = display_img - display_img.min()
                    if display_img_norm.max() > 0:
                        display_img_norm = display_img_norm / display_img_norm.max()
                else:
                    display_img_norm = np.clip(display_img, 0, 1)
            
            im = ax.imshow(display_img_norm)
        else:
            # Â§öÂÖâË∞±ÔºåÈÄâÊã©Ââç3‰∏™ÈÄöÈÅì‰Ωú‰∏∫RGB
            if n_channels >= 3:
                display_img = img[:3].transpose(1, 2, 0)  # ÂèñÂâç3ÈÄöÈÅì
            else:
                display_img = img[0]  # Âè™ÊòæÁ§∫Á¨¨‰∏ÄÈÄöÈÅì
            
            # ÂΩí‰∏ÄÂåñÊòæÁ§∫
            if display_img.max() > 1.1:
                display_img_norm = np.clip(display_img / 255.0, 0, 1)
            else:
                if display_img.min() < 0:
                    display_img_norm = display_img - display_img.min()
                    if display_img_norm.max() > 0:
                        display_img_norm = display_img_norm / display_img_norm.max()
                else:
                    display_img_norm = np.clip(display_img, 0, 1)
            
            if len(display_img_norm.shape) == 3:
                im = ax.imshow(display_img_norm)
            else:
                im = ax.imshow(display_img_norm, cmap='viridis')
        
        # Ê∑ªÂä†Ê†áÈ¢òÂíå‰ø°ÊÅØ
        min_val = img.min()
        max_val = img.max()
        ax.set_title(f'Âõæ{i}: {range_types[i]}\n[{min_val:.3f}, {max_val:.3f}]', 
                    fontsize=10)
        ax.axis('off')
        
        # Ê†πÊçÆÊï∞ÊçÆËåÉÂõ¥Á±ªÂûãÊ∑ªÂä†ÂΩ©Ëâ≤ËæπÊ°Ü
        if "Ë¥üÂÄº" in range_types[i]:
            color = 'red'
        elif "[0,1]" in range_types[i]:
            color = 'green'
        elif "[0,255]" in range_types[i]:
            color = 'blue'
        else:
            color = 'orange'
        
        # Ê∑ªÂä†ÂΩ©Ëâ≤ËæπÊ°Ü
        rect = Rectangle((0, 0), width-1, height-1, linewidth=3, 
                        edgecolor=color, facecolor='none')
        ax.add_patch(rect)
        
        # Ê∑ªÂä†È¢úËâ≤Êù°
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    
    # ÈöêËóèÂ§ö‰ΩôÁöÑÂ≠êÂõæ
    for i in range(n_samples, len(axes)):
        axes[i].axis('off')
    
    plt.suptitle(f'{data_name} - ÊØèÂº†ÂõæÂèØËßÜÂåñ\nÁ∫¢Ê°Ü=ÊúâË¥üÂÄº, ÁªøÊ°Ü=[0,1], ËìùÊ°Ü=[0,255], Ê©ôÊ°Ü=ÂÖ∂‰ªñ', 
                fontsize=14)
    plt.tight_layout()
    plt.show()

def check_data_consistency(h5_path):
    """
    Ê£ÄÊü•Êï∞ÊçÆ‰∏ÄËá¥ÊÄßÈóÆÈ¢ò
    """
    print(f"\nüîç Êï∞ÊçÆ‰∏ÄËá¥ÊÄßÊ£ÄÊü•")
    print("="*60)
    
    with h5py.File(h5_path, 'r') as f:
        # Ê£ÄÊü•ÊâÄÊúâÊï∞ÊçÆÈõÜÁöÑÊï∞ÊçÆËåÉÂõ¥
        issues = []
        
        for key in f.keys():
            if isinstance(f[key], h5py.Dataset):
                data = f[key][:]
                if len(data.shape) == 4:
                    min_vals = []
                    max_vals = []
                    
                    for i in range(data.shape[0]):
                        min_vals.append(data[i].min())
                        max_vals.append(data[i].max())
                    
                    min_vals = np.array(min_vals)
                    max_vals = np.array(max_vals)
                    
                    # Ê£ÄÊü•ÊòØÂê¶ÊúâË¥üÂÄº
                    negative_count = np.sum(min_vals < 0)
                    if negative_count > 0:
                        issues.append(f"{key}: {negative_count}Âº†ÂõæÊúâË¥üÂÄº")
                    
                    # Ê£ÄÊü•Êï∞ÊçÆËåÉÂõ¥‰∏ç‰∏ÄËá¥
                    range_spread = max_vals.max() - min_vals.min()
                    if range_spread > 255:
                        issues.append(f"{key}: Êï∞ÊçÆËåÉÂõ¥ËøáÂ§ß ({min_vals.min():.3f} Âà∞ {max_vals.max():.3f})")
                    
                    # Ê£ÄÊü•Ê∑∑ÂêàÊï∞ÊçÆËåÉÂõ¥
                    has_0_1 = np.any((min_vals >= 0) & (max_vals <= 1.1))
                    has_0_255 = np.any(max_vals > 100)
                    if has_0_1 and has_0_255:
                        issues.append(f"{key}: Ê∑∑ÂêàÊï∞ÊçÆËåÉÂõ¥ (Êó¢Êúâ[0,1]ÂèàÊúâ[0,255]ËåÉÂõ¥)")
        
        if issues:
            print("‚ö†Ô∏è ÂèëÁé∞ÁöÑÈóÆÈ¢ò:")
            for issue in issues:
                print(f"  ‚Ä¢ {issue}")
        else:
            print("‚úÖ Êú™ÂèëÁé∞ÊòéÊòæÁöÑÊï∞ÊçÆ‰∏ÄËá¥ÊÄßÈóÆÈ¢ò")

# ‰∏ªÂàÜÊûêÂáΩÊï∞
def main():
    h5_path = '/data2/users/yujieliang/dataset/Harvard_calibrated.h5'
    
    print("üîç ÂºÄÂßãÂàÜÊûêÊµãËØïÊï∞ÊçÆÈõÜ")
    print("="*80)
    
    # ÂàÜÊûêÂπ∂ÂèØËßÜÂåñ
    analyze_and_visualize_test_dataset(h5_path)
    
    # Êï∞ÊçÆ‰∏ÄËá¥ÊÄßÊ£ÄÊü•
    check_data_consistency(h5_path)
    
    print("\n" + "="*80)
    print("üèÅ ÂàÜÊûêÂÆåÊàê!")

# ËøêË°åÂàÜÊûê
main()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from matplotlib.patches import Rectangle
from collections import Counter

def analyze_mat_files(mat_file):
    """
    ÂàÜÊûêÊåáÂÆöÊñá‰ª∂Â§π‰∏≠img1.matÂà∞img6.matÁöÑÊï∞ÊçÆËåÉÂõ¥
    """
    # print(f"üîç ÂàÜÊûêÊñá‰ª∂Â§π: {folder_path}")
    
    # if not os.path.exists(folder_path):
    #     print(f"‚ùå Êñá‰ª∂Â§π‰∏çÂ≠òÂú®: {folder_path}")
    #     return
    
    # Ê£ÄÊü•img1.matÂà∞img6.matÊñá‰ª∂
    # mat_files = []
    # for i in range(1, 7):
    #     mat_file = os.path.join(folder_path, f"img{i}.mat")
    #     if os.path.exists(mat_file):
    #         mat_files.append((i, mat_file))
    #     else:
    #         print(f"‚ö†Ô∏è Êñá‰ª∂‰∏çÂ≠òÂú®: img{i}.mat")
    
    # if not mat_files:
    #     print("‚ùå Êú™ÊâæÂà∞‰ªª‰Ωïimg*.matÊñá‰ª∂")
    #     return
    
    # print(f"üìä ÊâæÂà∞ {len(mat_files)} ‰∏™.matÊñá‰ª∂")
    
    # ÂàÜÊûêÊØè‰∏™.matÊñá‰ª∂
    # all_data_info = []
    

    # Âä†ËΩΩ.matÊñá‰ª∂
    mat_data = loadmat(mat_file)
    
    # ÊâæÂà∞ÂÆûÈôÖÁöÑÊï∞ÊçÆÂèòÈáèÔºàÊéíÈô§matlabÁöÑÂÖÉÊï∞ÊçÆÔºâ
    data_keys = [k for k in mat_data.keys() if not k.startswith('__')]
    print(f"Êï∞ÊçÆÈîÆ: {data_keys}")
    
    for key in data_keys:
        data = mat_data[key]
        print(f"\nÊï∞ÊçÆÂèòÈáè: {key}")
        print(f"Êï∞ÊçÆÂΩ¢Áä∂: {data.shape}")
        print(f"Êï∞ÊçÆÁ±ªÂûã: {data.dtype}")
        
        # ÂàÜÊûêÊï∞ÊçÆËåÉÂõ¥
        min_val = float(data.min())
        max_val = float(data.max())
        mean_val = float(data.mean())
        std_val = float(data.std())
        
        # Âà§Êñ≠Êï∞ÊçÆËåÉÂõ¥Á±ªÂûã
        range_type = classify_data_range(min_val, max_val)
        
        data_info = {
            'key': key,
            'shape': data.shape,
            'dtype': data.dtype,
            'min': min_val,
            'max': max_val,
            'mean': mean_val,
            'std': std_val,
            'range_type': range_type,
            'data': data
        }
        # all_data_info.append(data_info)
        
        print(f"Êï∞ÊçÆËåÉÂõ¥: [{min_val:.6f}, {max_val:.6f}]")
        print(f"ÂùáÂÄº: {mean_val:.6f}")
        print(f"Ê†áÂáÜÂ∑Æ: {std_val:.6f}")
        print(f"ËåÉÂõ¥Á±ªÂûã: {range_type}")

    
    # Ê±áÊÄªÂàÜÊûê
    print_summary_analysis(data_info)
    
    # ÂèØËßÜÂåñ
    visualize_mat_data(data_info)
    
    return data_info

def classify_data_range(min_val, max_val):
    """
    Ê†πÊçÆÊï∞ÊçÆËåÉÂõ¥ÂàÜÁ±ª
    """
    if min_val < 0:
        if max_val <= 1:
            return "ÊúâË¥üÂÄº[<0,‚â§1]"
        elif max_val <= 255:
            return "ÊúâË¥üÂÄº[<0,‚â§255]"
        else:
            return f"ÊúâË¥üÂÄº[<0,{max_val:.2f}]"
    elif min_val >= 0 and max_val <= 1.1:
        return "[0,1]ËåÉÂõ¥"
    elif min_val >= 0 and max_val <= 255:
        if max_val > 100:
            return "[0,255]ËåÉÂõ¥"
        else:
            return f"[0,{max_val:.2f}]"
    else:
        return f"ÂÖ∂‰ªñ[{min_val:.2f},{max_val:.2f}]"

def print_summary_analysis(all_data_info):
    """
    ÊâìÂç∞Ê±áÊÄªÂàÜÊûêÁªìÊûú
    """
    print(f"\n{'='*80}")
    print("üìä Ê±áÊÄªÂàÜÊûê")
    print(f"{'='*80}")
    
    # ÂàõÂª∫Ê±áÊÄªË°®Ê†º
    print(f"{'Êñá‰ª∂':<15} {'ÂèòÈáè':<10} {'ÂΩ¢Áä∂':<20} {'ÊúÄÂ∞èÂÄº':<12} {'ÊúÄÂ§ßÂÄº':<12} {'ËåÉÂõ¥Á±ªÂûã':<20}")
    print("-" * 100)
    
    range_types = []
    for info in all_data_info:
        print(f"{info['file']:<15} {info['key']:<10} {str(info['shape']):<20} "
              f"{info['min']:<12.6f} {info['max']:<12.6f} {info['range_type']:<20}")
        range_types.append(info['range_type'])
    
    # ÁªüËÆ°ËåÉÂõ¥Á±ªÂûã
    range_counter = Counter(range_types)
    print(f"\nüìà Êï∞ÊçÆËåÉÂõ¥Á±ªÂûãÁªüËÆ°:")
    for range_type, count in range_counter.items():
        print(f"  {range_type}: {count}‰∏™Êï∞ÊçÆÈõÜ ({count/len(all_data_info)*100:.1f}%)")

def visualize_mat_data(all_data_info):
    """
    ÂèØËßÜÂåñ.matÊñá‰ª∂Êï∞ÊçÆ
    """
    if not all_data_info:
        return
    
    print(f"\nüé® ÂèØËßÜÂåñÊï∞ÊçÆ")
    
    # ËøáÊª§Âá∫ÂèØ‰ª•ÂèØËßÜÂåñÁöÑÊï∞ÊçÆÔºà2DÊàñ3DÂõæÂÉèÊï∞ÊçÆÔºâ
    visualizable_data = []
    for info in all_data_info:
        data = info['data']
        if len(data.shape) >= 2:  # Ëá≥Â∞ëÊòØ2DÊï∞ÊçÆ
            visualizable_data.append(info)
    
    if not visualizable_data:
        print("‚ö†Ô∏è Ê≤°ÊúâÂèØËßÜÂåñÁöÑÂõæÂÉèÊï∞ÊçÆ")
        return
    
    # ËÆ°ÁÆóÁΩëÊ†ºÂ∏ÉÂ±Ä
    n_data = len(visualizable_data)
    cols = min(3, n_data)
    rows = (n_data + cols - 1) // cols
    
    fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*5))
    if rows == 1 and cols == 1:
        axes = [axes]
    elif rows == 1 or cols == 1:
        axes = axes.flatten()
    else:
        axes = axes.flatten()
    
    for i, info in enumerate(visualizable_data):
        if i >= len(axes):
            break
            
        ax = axes[i]
        data = info['data']
        
        # Ê†πÊçÆÊï∞ÊçÆÂΩ¢Áä∂ÈÄâÊã©ÂèØËßÜÂåñÊñπÂºè
        if len(data.shape) == 2:
            # 2DÊï∞ÊçÆÔºåÁÅ∞Â∫¶Âõæ
            display_img = data
            im = ax.imshow(display_img, cmap='viridis')
        elif len(data.shape) == 3:
            if data.shape[2] == 3:
                # RGBÂõæÂÉè (H, W, 3)
                display_img = prepare_image_for_display(data)
                im = ax.imshow(display_img)
            elif data.shape[2] == 1:
                # ÂçïÈÄöÈÅìÂõæÂÉè (H, W, 1)
                display_img = data[:, :, 0]
                im = ax.imshow(display_img, cmap='viridis')
            else:
                # Â§öÈÄöÈÅìÔºåÊòæÁ§∫Á¨¨‰∏Ä‰∏™ÈÄöÈÅì
                display_img = data[:, :, 0]
                im = ax.imshow(display_img, cmap='viridis')
        elif len(data.shape) == 4:
            # 4DÊï∞ÊçÆÔºåÂèñÁ¨¨‰∏Ä‰∏™Ê†∑Êú¨ÁöÑÁ¨¨‰∏Ä‰∏™ÈÄöÈÅì
            if data.shape[0] > 0:
                if data.shape[-1] == 3:  # RGB
                    display_img = prepare_image_for_display(data[0])
                    im = ax.imshow(display_img)
                else:
                    display_img = data[0, :, :, 0] if data.shape[-1] > 1 else data[0, :, :, 0]
                    im = ax.imshow(display_img, cmap='viridis')
            else:
                continue
        else:
            # ÂÖ∂‰ªñÂΩ¢Áä∂ÔºåË∑≥Ëøá
            ax.text(0.5, 0.5, f"Êó†Ê≥ïÊòæÁ§∫\nÂΩ¢Áä∂: {data.shape}", 
                   ha='center', va='center', transform=ax.transAxes)
            ax.set_title(f"{info['file']}\n{info['key']}")
            continue
        
        # Ê∑ªÂä†Ê†áÈ¢òÂíå‰ø°ÊÅØ
        ax.set_title(f"{info['file']}\n{info['key']}: {info['range_type']}\n"
                    f"[{info['min']:.3f}, {info['max']:.3f}]", fontsize=10)
        ax.axis('off')
        
        # Ê†πÊçÆÊï∞ÊçÆËåÉÂõ¥Á±ªÂûãÊ∑ªÂä†ÂΩ©Ëâ≤ËæπÊ°Ü
        color = get_color_for_range_type(info['range_type'])
        
        # Ê∑ªÂä†È¢úËâ≤Êù°
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    
    # ÈöêËóèÂ§ö‰ΩôÁöÑÂ≠êÂõæ
    for i in range(len(visualizable_data), len(axes)):
        axes[i].axis('off')
    
    plt.suptitle('MATÊñá‰ª∂Êï∞ÊçÆÂèØËßÜÂåñ\nÁ∫¢Ê°Ü=ÊúâË¥üÂÄº, ÁªøÊ°Ü=[0,1], ËìùÊ°Ü=[0,255], Ê©ôÊ°Ü=ÂÖ∂‰ªñ', 
                fontsize=14)
    plt.tight_layout()
    plt.show()

def prepare_image_for_display(img):
    """
    ÂáÜÂ§áÂõæÂÉèÁî®‰∫éÊòæÁ§∫ÔºàÂΩí‰∏ÄÂåñÂà∞[0,1]ËåÉÂõ¥Ôºâ
    """
    if img.max() > 1.1:
        # ÂèØËÉΩÊòØ[0,255]ËåÉÂõ¥
        display_img = np.clip(img / 255.0, 0, 1)
    else:
        # ÂèØËÉΩÊòØ[0,1]ËåÉÂõ¥ÊàñÊúâË¥üÂÄº
        if img.min() < 0:
            # ÊúâË¥üÂÄºÔºåÂÖàÂπ≥ÁßªÂÜçÂΩí‰∏ÄÂåñ
            display_img = img - img.min()
            if display_img.max() > 0:
                display_img = display_img / display_img.max()
        else:
            display_img = np.clip(img, 0, 1)
    
    return display_img

def get_color_for_range_type(range_type):
    """
    Ê†πÊçÆÊï∞ÊçÆËåÉÂõ¥Á±ªÂûãËøîÂõûÈ¢úËâ≤
    """
    if "Ë¥üÂÄº" in range_type:
        return 'red'
    elif "[0,1]" in range_type:
        return 'green'
    elif "[0,255]" in range_type:
        return 'blue'
    else:
        return 'orange'

def main():
    """
    ‰∏ªÂáΩÊï∞
    """
    mat_file = '/data2/users/yujieliang/exps/Efficient-MIF-back-master-6-feat/data/HyperSpectral-Collections/allmats/Botswana.mat'
    
    print("üîç ÂºÄÂßãÂàÜÊûêCZ_hsdbÊñá‰ª∂Â§π‰∏≠ÁöÑMATÊñá‰ª∂")
    print("="*80)
    
    # ÂàÜÊûêMATÊñá‰ª∂
    all_data_info = analyze_mat_files(mat_file)
    
    print("\n" + "="*80)
    print("üèÅ ÂàÜÊûêÂÆåÊàê!")
    
    return all_data_info

# ËøêË°åÂàÜÊûê
if __name__ == "__main__":
    data_info = main()
```
