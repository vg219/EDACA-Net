```python
import numpy as np
import PIL.Image as Image
from pathlib import Path


over_imgs_dir = Path('/Data3/cao/ZiHanCao/datasets/MFF-RealMFF/NEAR')
under_imgs_dir = Path('/Data3/cao/ZiHanCao/datasets/MFF-RealMFF/FAR')
over_img_lst, under_img_lst = list(over_imgs_dir.glob('*')), list(under_imgs_dir.glob('*'))
sort_key = lambda x: x.stem
over_img_lst.sort(key=sort_key), under_img_lst.sort(key=sort_key)
stratedy = 'mean'

for over_p, under_p in zip(over_img_lst, under_img_lst):
    print(over_p.name, under_p.name)
    over, under = Image.open(over_p.as_posix()), Image.open(under_p.as_posix())
    over = over.convert('RGB')
    under = under.convert('RGB')
    over_arr, under_arr = np.array(over, dtype=np.float32), np.array(under, dtype=np.float32)
    if stratedy =='mean':
        result_arr = (over_arr + under_arr) / 2
    elif stratedy =='max':
        result_arr = np.maximum(over_arr, under_arr)
    result_arr = result_arr.astype(np.uint8)
    result = Image.fromarray(result_arr)
    result.save(f'/Data3/cao/ZiHanCao/datasets/MFF-RealMFF/MEAN/{over_p.name}')
    
```
